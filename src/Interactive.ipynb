```python
#Check if binary tree is full or not(All nodes except leaves have exactly two children)
def is_full_bt(tree):

    if((2*tree.num_leaves())==tree.root()+2):
        print("Binary Tree is full\n")
    else:
        print("Binary Tree is not full\n")

#Find nodes with exactly one child in binary trees which are not full
def nodes_with_one_child(tree):
    for node in tree.leaves_to_root_iterator(include_leaves = False, include_root = True):
        try:
            left = tree.child(0, node)
        except:
            print(node)

        try:
            right = tree.child(1, node)
        except:
            print(node)

def is_tree_binary(tree):
    for node in tree.leaves_to_root_iterator(include_leaves = False, include_root = True):
        try:
            tree.child(2,node)
            print(node)
            return False
    
        except:
            continue
    
    return True
```

```python
leaves = []
    
#returns the list of leaves in the subtree of the node
def find_leaves (node,tree):
    if(not tree.is_leaf(node)):
        #Left Subtree
        find_leaves(tree.child(0,node),tree)

        #Right Subtree
        find_leaves(tree.child(1,node),tree)
    
    #if leaf
    else:
        leaves.append(node)
```

```python
#returns the list of leaves in the subtree of the node
def num_leaves_node(node,tree):
#     if leaf
    if(tree.is_leaf(node)):
        return 1

    else:
        return num_leaves_node(tree.child(0,node),tree) + num_leaves_node(tree.child(1,node),tree)
```

```python
#returns the list of leaves for a list of nodes
def num_leaves_list(tree):
    
    leaf_per_node = np.zeros((tree.root()+1, ), dtype=int)
    
    for i in range(tree.root()+1):
        leaf_per_node[i] = num_leaves_node(i, tree)
    
    return leaf_per_node
```

```python
#gives the list of leaves belonging to a node in the unfiltered-tree
def filtered_node_region(filtered_node, original_tree, node_map):

    original_node = node_map[filtered_node]
    
    leaves.clear()
    find_leaves(original_node, original_tree)
    
    return leaves
```

```python
#gives the list of leaves belonging to a node in the filtered binary tree
def filtered_bin_node(filtered_bin_node, original_tree, filtered_bin_tree, new_nodes, node_map, node_map_bin):

    if filtered_bin_node not in new_nodes:
        filtered_node = node_map_bin[filtered_bin_node]
        return filtered_node_region(filtered_node, original_tree, node_map)
    
    #When chosen node is the transformed node
    child1, child2 = node_map_bin[filtered_bin_tree.child(0, filtered_bin_node)], node_map_bin[filtered_bin_tree.child(1, filtered_bin_node)]

    region_child1 = filtered_node_region(child1, original_tree, node_map)
    region_child2 = filtered_node_region(child2, original_tree, node_map)

    region_child1.extend(region_child2)
    
    return region_child1
```

```python
#gives the list of leaves belonging to a set of nodes in the filtered binary tree
def filtered_bin_region(G, new_nodes, original_tree, filtered_bin_tree, node_map, node_map_bin):
    
    region_leaves = []
    
    for node in G:
        region_leaves.extend(filtered_bin_node(node, original_tree, filtered_bin_tree, new_nodes, node_map, node_map_bin))
    
    return np.unique(region_leaves)    
```

```python
# Min Threshold for filtering BPT
FILTER_THRESHOLD = 1.5

print("Processing HED+SLIC Tree.............\n")

################ LOAD HED GRADIENT IMAGE #################
gradient = cv2.imread(f'../HED/hed_{IMAGE_NAME}.png', cv2.IMREAD_GRAYSCALE)

gradient = gradient.astype(np.float64)/255 
size = gradient.shape[:2]
imshow(gradient)

################ LOAD ~4000 SLIC SUPERPIXELS #################
slic = np.loadtxt(f'../SLIC/{IMAGE_NAME}.npy')


# Construct 4-adjacency graph of image
graph = hg.get_4_adjacency_graph(size)
edge_weights = hg.weight_graph(graph, gradient.astype(np.float32), hg.WeightFunction.mean)

# Compute the associated graph from slic label matrix
rag = hg.make_region_adjacency_graph_from_labelisation(graph, slic)

rag_edge_weights = hg.rag_accumulate_on_edges(rag, hg.Accumulators.min, edge_weights)


################ CONSTRUCT SINGLE LINKAGE HED BPT  #################
tree_slic_hed, altitudes_slic_hed = hg.binary_partition_tree_single_linkage(rag, rag_edge_weights)

print("HED+SLIC Tree constructed\n")

print("Number of nodes in Tree : %d"% tree_slic_hed.num_vertices())
print("Number of leaves in Tree : %d"% tree_slic_hed.num_leaves())

avg_node_area = np.mean(hg.attribute_area(tree_slic_hed)[:tree_slic_hed.num_leaves()])

print("Average size of leaf nodes in Tree : %0.1f\n\n"% avg_node_area)

################ FILTER HED BPT BY NODE AREA  ##################
filtered_tree_temp, filtered_altitudes_temp = hg.filter_small_nodes_from_tree(tree_slic_hed, altitudes_slic_hed, FILTER_THRESHOLD*avg_node_area)
filtered_tree, node_map = hg.simplify_tree(filtered_tree_temp, filtered_altitudes_temp == filtered_altitudes_temp[filtered_tree_temp.parents()], process_leaves=True)
filtered_altitudes = filtered_altitudes_temp[node_map]


################ CONSTRUCTED BINARY TREE FROM FILTERED TREE  ##################
filtered_tree_bin, node_map_new = hg.tree_2_binary_tree(filtered_tree)
filtered_altitudes_bin = filtered_altitudes[node_map_new]

print("Number of nodes in Filtered Tree : %d"% filtered_tree_bin.num_vertices())
print("Number of leaves in  Filtered Tree : %d\n"% filtered_tree_bin.num_leaves())

# Saliency map of filtered tree
imshow(hg.graph_4_adjacency_2_khalimsky(graph, hg.saliency(filtered_tree_temp, filtered_altitudes_temp, handle_rag=True)))
```

```python
#Returns intersection of leaf nodes of tree with the constructed Ground Truth foreground object
def leaf_intersection_with_gt(tree, gt_image, label_matrix):
    gt_inter_leaf = np.zeros((tree.num_leaves(), ))

    for leaf in tree.leaves():
        indices = np.where(label_matrix==float(leaf))
        inter_area = 0
        
        for i in range(len(indices[0])):
            if (gt_image[indices[0][i]][indices[1][i]]==1):
                inter_area+=1
        
        gt_inter_leaf[leaf] = inter_area

    return gt_inter_leaf
```

```python
#returns an array with value at each index(node value) as intersection of node with ground truth for trees which use superpixel nodes
def intersection_sp_nodes(tree, gt_image, label_matrix):
    gt_input = leaf_intersection_with_gt(tree, gt_image, label_matrix)
    
    return (hg.accumulate_sequential(tree, gt_input, hg.Accumulators.sum)) 
```

```python
#returns an array with value at each index(node value) as intersection of node with ground truth for trees which use superpixel nodes
def intersection_sp_nodes_filtered(tree, gt_image, leaf_intersection):
    
    intersection_nodes = np.zeros((tree.root()+1), dtype=int)
    
    for node in tree.leaves_to_root_iterator(include_leaves = True, include_root = True):
        if (tree.is_leaf(node)):
            intersection_nodes[node] = leaf_intersection[node]
        else:
            find_leaves(node, tree)
            inter = 0
            for leaf in leaves:
                inter+=leaf_intersection[leaf]
            intersection_nodes[node] = inter
            leaves.clear()
            
    return intersection_nodes
```
